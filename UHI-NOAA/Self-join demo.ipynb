```python
import pandas as pd
```

```python
stations = pd.DataFrame(['A','A','A','B','B','B','B','C','C','C','C'], columns = ['station'])
stations['date'] = ['2019-01-01','2019-01-01',
                    '2019-01-02',
                    '2019-01-01','2019-01-01',
                    '2019-01-02','2019-01-02',
                    '2019-01-01','2019-01-01',
                    '2019-01-02','2019-01-02']
stations['type'] = ['min','max','min','min','max','min','max','min','max','min','max',]
stations['temp'] = [1,10,2,3,30,4,40,5,50,6,60] 
stations['NN']   = ['B','B','B','A','A','A','A','A','A','A','A']
stations
```

```python
stations = stations.merge(stations, left_on=['NN','date','type'], right_on=['station','date','type'])
stations
```

Clean up:



```python
stations.drop(['NN_y', 'station_y'], axis=1, inplace=True)
stations.rename(index=str, columns={"station_x": "station ", "NN_x": "NN", "temp_x": "temp", "temp_y": "NN_temp"}, inplace=True)
stations
```
